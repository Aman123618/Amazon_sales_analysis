```python
ax=sns.countplot(x='Size',data=df) #This show only graph
plt.ylabel('Sale')
for bar in ax.containers:
    ax.bar_label(bar)    #This also label  the bars
```

```python
df.info()
```

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by=['Qty'],ascending=False)
```

```python
s_qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by=['Qty'],ascending=False)


s=sns.barplot(x='Size',y='Qty',data=s_qty)
```

Courier Status

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Courier Status',data=df,hue='Status')
plt.show()
```

Note: majority of order are delivered to buyer

```python
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=25,edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

Note: From above graph we can see that T-shirt are most selling

```python
B2B_check=df['B2B'].value_counts()
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
plt.show()
```

Note: From above graph we can say that 99.2% buyers are retailers and 0.8% are B2B

```python
x_data=df['Category']
y_data=df['Size']
plt.scatter(x_data,y_data)
plt.show
```

From above chart we can see the availabity of sizes in diff product


```python
df.info()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('counts')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
#top 10 state
top_10_state=df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('counts')
plt.title('Distribution of state')
plt.xticks(rotation=45)
plt.show()
```

Note : Above chart shows that most of the buyers are from Maharashtra
